```python
class_1 = df[df['fetal_health'] == 1]
class_2 = df[df['fetal_health'] == 2]
class_3 = df[df['fetal_health'] == 3]
```

```python
size = int(0.75 * len(class_1))

test_class_1 = class_1[size:]
size = int(0.75 * len(class_2))

test_class_2 = class_2[size:]
size = int(0.75 * len(class_3))

test_class_3 = class_3[size:]
```
